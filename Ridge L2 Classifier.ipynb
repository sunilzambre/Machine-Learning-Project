```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as matlab
import warnings
warnings.filterwarnings("default")
```

```python
ckd_df =pd.read_csv("CKD.csv")
```

```python
ckd_df = pd.get_dummies(ckd_df,drop_first=True)
```

```python
independent=ckd_df[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
```

```python
independent
```

```python
independent.shape
```

```python
dependent = ckd_df[['classification_yes']]
```

```python
dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=1/3,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
from sklearn.linear_model import RidgeClassifier
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={'alpha':[1],"fit_intercept": [True], "solver": ['cholesky']}
grid=GridSearchCV(RidgeClassifier(),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
#fitting the model for grid search
grid.fit(X_train,Y_train)
```

```python
from sklearn.linear_model import RidgeClassifier
classifier=RidgeClassifier(random_state=0)
classifier.fit(X_train,Y_train)
```

```python
y_pred=classifier.predict(X_test)
y_pred
```

```python
re=grid.cv_results_
grid_predictions=grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(Y_test,y_pred)
```

```python
print(cm)
```

```python
from sklearn.metrics import classification_report
clf_report=classification_report(Y_test,y_pred)
```
