```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_circles
from sklearn.neighbors import LocalOutlierFactor # Algorithm for Anomaly detection
```

```python
X ,_ = make_circles(n_samples=1000,noise=0.11,random_state=42,factor=0.3)
```

### Visualizing data before Applying Algorithm

```python
plt.scatter(x=X[:,0],y=X[:,1])
```

```python
lof = LocalOutlierFactor(n_neighbors=10,algorithm='auto',contamination='auto',n_jobs=-1)
lof.fit_predict(X=X)
```

```python
lof.get_params()
```

```python
plt.scatter(x=X[:,0],y=X[:,1],c=lof.fit_predict(X))
```

```python
lst = []
for outlier in lof.fit_predict(X=X):
    if outlier == -1:
        lst.append(outlier)
    else:
        pass
```

```python
print(f'I have only {len(lst)} Outliers in my dataset')
```
