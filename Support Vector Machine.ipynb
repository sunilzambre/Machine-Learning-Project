## Support Vector Machine

- Support Vector Machine (SVM) are a popular supervised machine learning algorithm used for classification and regression     tasks. SVMs are particularly effective when dealing with complex datasets and are widely used in various domains,           including text classification, image classification, and bioinformatics.

- The primary objective of an SVM is to find an optimal hyperplane that separates the data points into different classes.     The hyperplane is selected in such a way that it maximizes the margin, which is the distance between the hyperplane and     the nearest data points from each class. These data points, known as support vectors, play a crucial role in determining   the hyperplane.

## Support Vector Machine (SVM)

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['target','flower_name'], axis='columns')
y = df.target
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.7)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict([[5.2,3.1,2.4,1.5]])
```

```python
model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

## Tuning Parameters

## Regularization(C)

```python
model_reg = SVC(C=1)
model_reg.fit(X_train,y_train)
model_reg.score(X_test,y_test)
```

```python
model_reg = SVC(C=7)
model_reg.fit(X_train,y_train)
model_reg.score(X_test,y_test)
```

## Gamma

```python
model_G = SVC(gamma=1)
model_G.fit(X_train,y_train)
model_G.score(X_test,y_test)
```

```python
model_G = SVC(gamma=5)
model_G.fit(X_train,y_train)
model_G.score(X_test,y_test)
```

## Kernel

```python
model_kernel = SVC(kernel='linear')
model_kernel.fit(X_train,y_train)
model_kernel.score(X_test,y_test)
```
